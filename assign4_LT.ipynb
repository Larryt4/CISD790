Parallel coordinates are used to present a dataset of records with *n* quantitative attributes. We define *n* vertical, evenly-spaced axes, one per dimension. The range of axis *u* extends from the minimum value for attribute *i* (at the bottom) up to its maximum value. A point *(x1, x2,…,xn)* is represented by a polyline with a vertex on each axis. The polyline passes through the vertex positioned at *xi* along axis *i*, for each *i*.

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None
```

```python
# parallel coordinates supported by pandas
from pandas.tools.plotting import parallel_coordinates
parallel_coordinates(data, 'Species')
```

In the figure above, each polyline represents a record in this 4-column dataset; alternatively, each is a point in 4-dimensional attribute space. Each polyline crosses a vertical axis at a point corresponding to its value in that dimension.

One problem with the above version of parallel coordinates is that the different numeric columns (dimensions) are plotted *on the same scale*, even though their ranges vary. In this example, we see that the range of Sepal Length is [4.3, 7.9], whereas the range of Petal Width is [0.1, 2.5], as seen below, and yet every dimension has the same range [0,8].

```python
data.min(), data.max()
```

Since attributes are often not normalized to a common range, some of the dimensions are compressed in the resulting parallel coordinates (Petal Width in this example). The problem is not so pronounced here because the ranges across the dimensions are not so different. But suppose we were to scale the Sepal Length dimension by a factor of 10 and the Petal Width dimension by a factor of 0.1.

```python
data2 = data.ix[:,:-1] * [10.0, 1.0, 1.0, 0.1]  # broadcast down first 4 columns of data
data2['Species'] = data['Species']
parallel_coordinates(data2, 'Species')
```

In this visualization of data2, we can barely distinguish the records' values in the Petal Width attribute. A more robust approach is to scale each dimension independently to its range [min, max], where min and max are the minimum and maximum values in that dimension. In the resulting parallel coordinates, we label the range of each axis independently.

```python

def parallel_coordinates(data, class_column, cols=None, ax=None, colors=None,
                         use_columns=False, xticks=None, colormap=None, **kwds):
    """Parallel coordinates plotting.

    Parameters
    ----------
    data: DataFrame
        A DataFrame containing data to be plotted
    class_column: str
        Column name containing class names
    cols: list, optional
        A list of column names to use
    ax: matplotlib.axis, optional
        matplotlib axis object
    colors: list or tuple, optional
        Colors to use for the different classes
    use_columns: bool, optional
        If true, columns will be used as xticks
    xticks: list or tuple, optional
        A list of values to use for xticks
    colormap: str or matplotlib colormap, default None
        Colormap to use for line colors.
    kwds: list, optional
        A list of keywords for matplotlib plot method

    Returns
    -------
    ax: matplotlib axis object

    Examples
    --------
    >>> from pandas import read_csv
    >>> from pandas.tools.plotting import parallel_coordinates
    >>> from matplotlib import pyplot as plt
    >>> df = read_csv('https://raw.github.com/pydata/pandas/master/pandas/tests/data/iris.csv')
    >>> parallel_coordinates(df, 'Name', colors=('#556270', '#4ECDC4', '#C7F464'))
    >>> plt.show()

    Source
    --------
    https://github.com/pydata/pandas/blob/master/pandas/tools/plotting.py
    """
    from pandas.tools.plotting import _get_standard_colors


    n = len(data)
    classes = set(data[class_column])
    class_col = data[class_column]

    if cols is None:
        df = data.drop(class_column, axis=1)
    else:
        df = data[cols]

    used_legends = set([])

    ncols = len(df.columns)

    # determine values to use for xticks
    if use_columns is True:
        if not np.all(np.isreal(list(df.columns))):
            raise ValueError('Columns must be numeric to be used as xticks')
        x = df.columns
    elif xticks is not None:
        if not np.all(np.isreal(xticks)):
            raise ValueError('xticks specified must be numeric')
        elif len(xticks) != ncols:
            raise ValueError('Length of xticks must match number of columns')
        x = xticks
    else:
        x = range(ncols)

    if ax is None:
        ax = plt.gca()

    color_values = _get_standard_colors(num_colors=len(classes),
                                        colormap=colormap, color_type='random',
                                        color=colors)

    colors = dict(zip(classes, color_values))

    for i in range(n):
        row = df.irow(i).values
        y = row
        kls = class_col.iget_value(i)
        if kls not in used_legends:
            label = kls
            used_legends.add(label)
            ax.plot(x, y, color=colors[kls],
                    label=label, **kwds)
        else:
            ax.plot(x, y, color=colors[kls], **kwds)

    for i in x:
        ax.axvline(i, linewidth=1, color='black')

    ax.set_xticks(x)
    ax.set_xticklabels(df.columns)
    ax.set_xlim(x[0], x[-1])
    ax.legend(loc='upper right')
    ax.grid()
    return ax
```

```python
def parallel_coordinates2(data, class_column,cols=None, colors=None, colormap=None, **kwds):
    
    from pandas.tools.plotting import _get_standard_colors
    

    n = len(data)
    classes = set(data[class_column])
    class_col = data[class_column]
    if cols is None:
        #move class_col to end of data
        data = data.drop(class_column, axis=1)
        data['class_column'] = class_col
    else:
        data = data[cols]
        data['class_column'] = class_col
    
    columnNames = data.columns.values.tolist()
    
    color_values = _get_standard_colors(num_colors=len(classes),colormap=colormap, color_type='random', color=colors)
    colors = dict(zip(classes, color_values))
    used_legends = set([])
    used_lines = []
    ##set up subplots for each set of points
    fig, axes = plt.subplots(1, len(data.columns)-2, sharey=False,figsize=(14,10))
    
    
    #get min,max,range for each column and normalize data
    minVals = []
    maxVals = []
    for i in range(len(data.columns)-1):
        mn = min(data[data.columns[i]])
        mx = max(data[data.columns[i]])
        r = mx-mn
        data[data.columns[i]] = (data[data.columns[i]]-mn)/r
        minVals.append(mn)
        maxVals.append(mx)
    
    x = range(2)
    first = True
    for i, ax in enumerate(axes):
        df=data[data.columns[i:i+2]]
        ax.set_ylim([0.0,1.0])
        ax.set_yticks([])
        ax2 = ax.twiny()
        if first:
            ax.set_xticks([0.0,1.0], minor=True)
            ax.set_xticks([0.0,1.0], minor=False)
            ax.set_xticklabels([minVals[i],minVals[i+1]],fontsize=14, color='blue', minor=True)
            ax.set_xticklabels(['\n'+columnNames[i],'\n'+columnNames[i+1]],fontsize=16, color='black',weight='bold',minor=False)
            ax2.set_xticks([0.0,1.0])
            ax2.set_xticklabels([maxVals[i],maxVals[i+1]],fontsize=14, color='blue')
            first=False
        else:
            ax.set_xticks([1.0], minor=True)
            ax.set_xticks([1.0], minor=False)
            ax.set_xticklabels([minVals[i+1]],fontsize=14, color='blue', minor=True)
            ax.set_xticklabels(['\n'+columnNames[i+1]],fontsize=16, color='black',weight='bold', minor=False)
            ax2.set_xticks([1.0])
            ax2.set_xticklabels([maxVals[i+1]],fontsize=14, color='blue')
        for j in range(n):
            row = df.irow(j).values
            y = row
            kls = class_col.iget_value(j)
            if kls not in used_legends:
                label = kls
                used_legends.add(label)
                l1, = ax.plot(x, y, color=colors[kls], label=label,linewidth=2.0, **kwds)
                used_lines.append(l1)
            else:
                ax.plot(x, y, color=colors[kls],linewidth=2.0, **kwds)
        
        
    plt.figlegend(used_lines,used_legends,loc='upper right',borderaxespad=4.0,fontsize=16)
    plt.subplots_adjust(wspace=0)
    plt.show()

parallel_coordinates2(data, 'Species')
```

parallel_coordinates3 below adds the ability to show a line representing the average of each class.  The average is shown by default.

```python
def parallel_coordinates3(data, class_column,cols=None, colors=None, colormap=None, show_avg=True, **kwds):
    
    from pandas.tools.plotting import _get_standard_colors

    n = len(data)
    classes = set(data[class_column])
    class_col = data[class_column]
    if cols is None:
        #move class_col to end of data
        data = data.drop(class_column, axis=1)
        data['class_column'] = class_col
    else:
        data = data[cols]
        data['class_column'] = class_col
    
    columnNames = data.columns.values.tolist()
    
    color_values = _get_standard_colors(num_colors=len(classes),colormap=colormap, color_type='random', color=colors)
    colors = dict(zip(classes, color_values))
    used_legends = set([])
    used_lines = []
    ##set up subplots for each set of points
    fig, axes = plt.subplots(1, len(data.columns)-2, sharey=False,figsize=(14,10))
    
    
    #get min,max,range for each column and normalize data
    minVals = []
    maxVals = []
    for i in range(len(data.columns)-1):
        mn = min(data[data.columns[i]])
        mx = max(data[data.columns[i]])
        r = mx-mn
        data[data.columns[i]] = (data[data.columns[i]]-mn)/r
        minVals.append(mn)
        maxVals.append(mx)
      
    #get values for plotting avg lines
    if show_avg:
        avgs = [] 
        tempClasses= np.array(list(classes))
        for i in range(len(classes)):
            tempDF = data[data['class_column']==tempClasses[i]]
            avgs.append(list(tempDF.mean(axis=0, numeric_only=True).values))
        avgDF = pd.DataFrame.from_dict(avgs)
    
    
    x = range(2)
    first = True
    for i, ax in enumerate(axes):
        df=data[data.columns[i:i+2]]
        ax.set_ylim([0.0,1.0])
        ax.set_yticks([])
        ax2 = ax.twiny()
        if first:
            ax.set_xticks([0.0,1.0], minor=True)
            ax.set_xticks([0.0,1.0], minor=False)
            ax.set_xticklabels([minVals[i],minVals[i+1]],fontsize=14, color='blue', minor=True)
            ax.set_xticklabels(['\n'+columnNames[i],'\n'+columnNames[i+1]],fontsize=16, color='black',weight='bold',minor=False)
            ax2.set_xticks([0.0,1.0])
            ax2.set_xticklabels([maxVals[i],maxVals[i+1]],fontsize=14, color='blue')
            first=False
        else:
            ax.set_xticks([1.0], minor=True)
            ax.set_xticks([1.0], minor=False)
            ax.set_xticklabels([minVals[i+1]],fontsize=14, color='blue', minor=True)
            ax.set_xticklabels(['\n'+columnNames[i+1]],fontsize=16, color='black',weight='bold', minor=False)
            ax2.set_xticks([1.0])
            ax2.set_xticklabels([maxVals[i+1]],fontsize=14, color='blue')
        for j in range(n):
            row = df.irow(j).values
            y = row
            kls = class_col.iget_value(j)
            if kls not in used_legends:
                label = kls
                used_legends.add(label)
                l1, = ax.plot(x, y, color=colors[kls], label=label,linewidth=2.0, **kwds)
                used_lines.append(l1)
            else:
                ax.plot(x, y, color=colors[kls],linewidth=2.0, **kwds)
        if show_avg:
            for k in range(len(avgDF)):
                avgdf=avgDF[avgDF.columns[i:i+2]]
                row = avgdf.irow(k).values
                y = row
                kls = tempClasses[k]
                ax.plot(x, y, color='black', linewidth=10.0, **kwds)
                ax.plot(x, y, color=colors[kls], linewidth=8.0, **kwds)
        
        
    plt.figlegend(used_lines,used_legends,loc='upper right',borderaxespad=4.0,fontsize=16)
    plt.subplots_adjust(wspace=0)
    plt.show()

parallel_coordinates3(data, 'Species', show_avg=True)
```
