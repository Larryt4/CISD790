## Section1: SOM

The SOM is based on the work of Kohonen (1982). First the helper functions for the SOM were defined.  These were adapted from ideas on implementing SOM in iPython found in this <a href='http://nbviewer.ipython.org/gist/anonymous/3407544'>notebook</a>.  

```python
#helper functions

#normalize the data
def normalizeDf(df):
    for i in range(len(df.columns)):
        mn = min(df[df.columns[i]])
        mx = max(df[df.columns[i]])
        r = mx-mn
        df[df.columns[i]] = (df[df.columns[i]]-mn)/r
    return df

#create SOM map node index
def create_output_nodes(num_nodes, map_size):
    output_nodes = np.zeros((num_nodes, 2), dtype=int)
    i=0
    k=0
    for n in range(num_nodes):
            output_nodes[n] = [i,k]
            k+=1
            if k == map_size:
                k=0
                i+=1
                if i == map_size:
                    i=0
    return output_nodes

#returns index of closest output node
def find_winner(weights, input_vector):
    winner = 0
    current_best_distance = sys.float_info.max
    num_nodes, num_dims = weights.shape
    for i in range(num_nodes):
        distance = 0.0
        for j in range(num_dims):
            distance += (((input_vector[j] - weights[i,j]) ** 2.0) ** 0.5)
        if distance < current_best_distance:
            current_best_distance = distance
            winner = i
    return winner


#update the network weights  
def update_weights(input_vector, weights, winner, output_nodes, learning_rate, neighborhood):
    
    winner_x = output_nodes[winner,0]
    winner_y = output_nodes[winner,1]
    error = 0.0
    
    num_nodes, num_dims = weights.shape
    for i in range(num_nodes):
        grid_distance = (((winner_x - output_nodes[i,0]) ** 2.0) + ((winner_y - output_nodes[i,1])**2.0))**0.5
        dampening = np.exp(-1.0 * grid_distance / ( 2.0 * neighborhood**2.0))
        dampening *= learning_rate
        
        for j in range(num_dims):
            delta = dampening * (input_vector[j] - weights[i,j])
            weights[i,j] += delta
            error += (delta**2)**0.5
    return error

def scale(start, end, position_between_zero_and_one):
    return float(start + position_between_zero_and_one * (end - start))
```

Next the main function for the SOM was defined.  It returns two result sets, the first is the coordinates of the training data and the second contains the coordinates of the test data.

```python
def run_SOM(data, unknown_class_data, map_size, num_iterations):
    data = normalizeDf(data)
    uc_data = normalizeDf(unknown_class_data)
    num_nodes = map_size*map_size
    num_dims = len(data.columns)
    output_nodes = create_output_nodes(num_nodes, map_size)
    #randomize weights
    weights = np.random.random((num_nodes, num_dims))
    
    l_rate_initial = 0.5
    l_rate_final = 0.1
    neighborhood_size_initial = map_size/2
    neighborhood_size_final = 1.0

    for it in range(num_iterations):
        cError = 0.0
        for i in range(len(data)):

            winner = find_winner(weights, data.iloc[i].values)
            
            # Calculate the new learning rate and new neighborhood
            normalized_progress = float(it) / float(num_iterations)
            l_rate = scale(l_rate_initial, l_rate_final, normalized_progress)
            neighborhood_size = scale(neighborhood_size_initial, neighborhood_size_final, normalized_progress)
            
            cError += update_weights(data.iloc[i].values, weights, winner, output_nodes, l_rate, neighborhood_size)
        #print cError #for testing 

    som_results = np.zeros((len(data),2),dtype=float)
    uk_results = np.zeros((len(uc_data),2),dtype=float)
    for k in range(len(data)):
        winner = find_winner(weights, data.iloc[k].values)
        winner_x = output_nodes[winner,0]
        winner_y = output_nodes[winner,1]
        som_results[k,0] = winner_x
        som_results[k,1] = winner_y
    for m in range(len(uc_data)):
        winner = find_winner(weights, uc_data.iloc[m].values)
        winner_x = output_nodes[winner,0]
        winner_y = output_nodes[winner,1]
        uk_results[m,0] = winner_x
        uk_results[m,1] = winner_y
    return som_results, uk_results
```

```python
som_results, uk_results = run_SOM(testData,unknownData, 20, 20) 
```

As shown the SOM clusters the data well with the centroids shown as square markers.  Next the distance of each unknown class point to the centroids was determined.  These distances were used to determine which class an unknown point was more likely to be a member of.

We can use a distance of 0.0 to the centroid to indicate 100% probability of membership in a class and the maximum distance on the map to represent 0%.  Scaling the actual distance to cenroid linearly yields the probability of membership in a class.  Results shown below.

```python
centroids, labels = vq.kmeans2(som_results, 2, minit='points')
def centroid_distance(centroids, point, map_size):
    max_distance = (map_size**2+map_size**2)**0.5
    results = np.zeros((len(centroids),3),dtype=float)
    for i in range (len(centroids)):
        results[i,0] = centroids[i,0]
        results[i,1] = centroids[i,1]
        distance = (((point[0] - centroids[i,0]) ** 2.0) + ((point[1] - centroids[i,1])**2.0))**0.5
        results[i,2] = (distance/max_distance -1)*(-1)
    return pd.DataFrame(results,columns=['cX', 'cY', 'P'])
```

## Section 2: polSOM

The polSOM is based on the work of Xu, Xu, & Chow (2010). Again first the helper functions were defined.

```python
def create_polar_benchmark_nodes(num_dims, num_tori):
    output_nodes = np.zeros((num_dims*num_tori, 2), dtype=float)
    angleCoords = np.zeros((num_dims,1), dtype=float)
    for i in range(len(angleCoords)):
        angleCoords[i] = 2*(np.pi/num_dims)*i
    r=1
    t=0
    for n in range(num_dims*num_tori):
            output_nodes[n] = [r,angleCoords[t]]
            t+=1
            if t == num_dims:
                t=0
                r+=1
                if r > num_tori:
                    break
    return output_nodes

```

The benchmark nodes were poltted to ensure they were positioned correctly.

```python
polNodes = create_polar_benchmark_nodes(6,2)
fig, axes = plt.subplots(figsize=(10,10))
ax = plt.subplot(111, polar=True)
for i in range(len(polNodes)):
    ax.scatter(polNodes[i,1], polNodes[i,0], color='b',s=80)
#axes.set_rmax(2.0)
ax.grid(True)

plt.show()
```

As shown the benchmark nodes divide the polar map into regions corresponding to each dimension of the data. 

```python
#scale the weights so that max weight is 1.0
#make primary angle value twice others
def set_benchmark_weights(num_dims, num_tori):
    num_nodes = num_dims*num_tori
    polWeights = np.zeros((num_nodes,num_dims),dtype=float)
    inc = 0.5/(num_tori)
    cInc = inc
    #start from 0 at lowest tori, same as node indices
    primary = 0
    for i in range(num_nodes):
        for k in range(num_dims):
            if k==primary:
                polWeights[i,k]=2*cInc
            else:
                polWeights[i,k]=cInc
        primary+=1
        if primary == num_dims:
            primary = 0
            cInc+=inc
    return polWeights
            
        
```

```python
#create polar coords array and set to random values within appropriate range
def set_data_polar_coords(num_data_points,num_tori):
    data_polar_coords = np.zeros((num_data_points,2),dtype=float)
    for i in range(num_data_points):
        data_polar_coords[i,0]=np.random.uniform(0,num_tori)
        data_polar_coords[i,1]=np.random.uniform(0,np.pi*2)
    return data_polar_coords
```

```python
def run_polSOM(data, num_tori, num_iterations):
    data = normalizeDf(data)

    num_dims = len(data.columns)
    num_nodes = num_dims*num_tori
    benchmark_nodes = create_polar_benchmark_nodes(num_dims,num_tori)
    polWeights = set_benchmark_weights(num_dims, num_tori)
    data_polar_coords = set_data_polar_coords(len(data),num_tori)

    l_rate_initial_1 = 0.4
    l_rate_final_1 = 0.1
    l_rate_initial_2 = 0.3
    l_rate_final_2 = 0.1
    neighborhood_size_initial = num_nodes/2
    neighborhood_size_final = 1.0

    for it in range(num_iterations):
        cError = 0.0
        for i in range(len(data)):

            winner = find_winner(polWeights, data.iloc[i].values)
            
            # Calculate the new learning rate and new neighborhood
            normalized_progress = float(it) / float(num_iterations)
            l_rate_1 = scale(l_rate_initial_1, l_rate_final_1, normalized_progress)
            l_rate_2 = scale(l_rate_initial_2, l_rate_final_2, normalized_progress)
            neighborhood_size = scale(neighborhood_size_initial, neighborhood_size_final, normalized_progress)
            pol_update_weights(i,data.iloc[i].values, polWeights, winner, benchmark_nodes, l_rate_1,l_rate_2, neighborhood_size, data_polar_coords)
    return data_polar_coords
    
```

polSOM displays data points based on the weight of the data (radius) and the features (angle). Since the features are the important classifier in this experiment, instead of using a kMeans centroid the median angle of each class can be used as the centroid.

```python
class1_mean = class1.iloc[:,1].mean(axis=1)
class2_mean = class2.iloc[:,1].mean(axis=1)
unknown['P1'] = ((2*np.pi - np.abs(unknown.iloc[:,1]-class1_mean))/(2*np.pi))
unknown['P2'] = ((2*np.pi - np.abs(unknown.iloc[:,1]-class2_mean))/(2*np.pi))
unknown
```

Similar to the method used for SOM above, equal to the mean angle for the class equates to 100% probabilty of membership and a difference of 180 equates to 0% probability of membership.  The table above shows the probablilty that each point is a member of class 1 (P1) or class 2 (P2).  These preliminary results indicate polSOM provides a more accurate determination of class membership. It can also be seen that since polSOM provides individual coordinates for each data point, visually the data points more accurately reflect the clustering of the data points.

As was seen with the test data, polSOM provides visually superior clustering of the data.  Also the primary hurdle of anomaly-based intrusion detection can be seen in both figures 3 and 4. Future work in this effort will be to attempt to reduce the number of false positives using post processing such as naive Bayes.
